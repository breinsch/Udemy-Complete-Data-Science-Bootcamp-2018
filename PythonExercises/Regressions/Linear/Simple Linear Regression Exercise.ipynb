## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
# seaborn improves on matplotlib's plots
import seaborn as sns
sns.set()
```

## Load the data

```python
data = pd.read_csv('real_estate_price_size.csv')
print(data.describe())
data
```

## Create the regression

### Declare the dependent and the independent variables

```python
y = data['price']
x_1 = data['size'] # using y = x_0b_0 + x_1b_1
```

### Explore the data

```python
plt.scatter(x_1, y)
plt.xlabel('Size', fontsize=20)
plt.ylabel('Price', fontsize=20)
plt.show()
```

### Regression itself

```python
# we create vector x with its components x_0 = 1 and x_1 = ind. var.
x = sm.add_constant(x_1)
# perform regression using ordinary least squares on the error
results = sm.OLS(y, x).fit()
# display regression results
results.summary()
```

### Plot the regression line on the initial scatter

```python
# a note on the above table: t and P>|t| are the t statistic and
# its corresponding p value when the null hypothesis is that b_i = 0
# so as we can see, since the p value is the probability that H_0
# is correct, then we can reject the H_0. also R^2 is a measure that
# means (SUM_i y_hat_i - y_bar)/(SUM_i y_i - y_bar), so it gives
# a value between 0 and 1 where 1 is a great fit and 0 not so much,
# i.e. a measure of explanation of variablity
plt.scatter(x_1, y)
# prediction y_hat = b_0x_0 + b_1x_1 constants taken from coef above
y_hat = (1.019e+05 * 1) + (223.1787 * x_1)
fig = plt.plot(x_1, y_hat, lw=4, c='orange', label="regression line")
plt.xlabel('Size', fontsize=20)
plt.ylabel('Price', fontsize=20)
plt.show()
```
